```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv("student_data.csv")
df
```

```python
sns.boxplot(x="sex",y="G3",data=df)
plt.title("gender vs grade")
plt.show()
```

```python
sns.scatterplot(x="G2",y="G3",data=df)
plt.title("Grade2 vs Grade3")
plt.show()
```

```python
sns.barplot(x='Walc', y='G3', data=df)
plt.title("Weekend Alcohol Consumption vs Grade")
plt.show()
```

```python
sns.scatterplot(x='absences', y='G3', data=df)
plt.title("Absences vs Final Grade")
plt.show()
```

```python
sns.barplot(x='Medu', y='G3', data=df)
plt.title("Mother Education vs Final Grade")
plt.show()
```

```python
sns.barplot(x='famrel', y='G3', data=df)
plt.title("Family Relationship vs Grade")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['G1','G2','G3','absences','Walc','Dalc','health']].corr(),
            annot=True, cmap='coolwarm')
plt.show()

```
